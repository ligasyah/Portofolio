```python
import pandas as pd
```

```python
df = pd.read_csv('olist_orders_dataset.csv')
```

```python
df
```

```python
#melihat tipe data column
df.info()

#melihat nama column di data
df.columns

#melihat salah satu colum
df['order_purchase_timestamp']
```

```python
#merubah dtype column date menjadi date time
df['order_purchase_timestamp']= pd.to_datetime(df['order_purchase_timestamp'])
```

```python
df.info()
```

```python
df.columns
```

```python
kolom_date = ['order_approved_at', 'order_delivered_carrier_date',
       'order_delivered_customer_date', 'order_estimated_delivery_date']
```

```python
#rumah dtype per column, gunakan for loop
for i in kolom_date:
  df[i]=pd.to_datetime(df[i])
```

```python
df.info()
```

```python
#menambahkan column tahun transaksi
df['Tahun Transaksi'] = df['order_purchase_timestamp'].dt.year

#menambahkan column bulan transakti
df['Bulan Transaksi'] = df['order_purchase_timestamp'].dt.month
#kl mau dirubah angka bulan jd nama, "dt.month_name()"
```

```python
df.head()
```

```python
#pivot per thun
df.groupby(['Tahun Transaksi']).agg({'order_id':'count'})
```

```python
df.groupby(['Tahun Transaksi','Bulan Transaksi']).agg({'order_id':'count'})
```

```python
df.groupby(['Tahun Transaksi','Bulan Transaksi']).agg({'order_id':'count'}).unstack()
```

```python
#rename duluu sblm di download
file_excel = df.groupby(['Tahun Transaksi','Bulan Transaksi']).agg({'order_id':'count'}).unstack()
```
